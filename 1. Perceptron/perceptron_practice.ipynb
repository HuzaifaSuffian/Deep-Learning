```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_classes=2,n_clusters_per_class=2,hypercube=False,random_state=12,class_sep=10)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1], c=y, cmap='coolwarm')
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(X,y)
```

```python
print('coef_ ',p.coef_)
print('intercept_ ',p.intercept_)
```

```python
m = -(p.coef_[0][0]/p.coef_[0][1])
b = -(p.intercept_/p.coef_[0][1])
```

```python
x_axis = np.linspace(-3,3,100)
y_axis = m*x_axis + b
```

```python
figure = plt.figure()
plt.plot(x_axis,y_axis)
plt.scatter(X[:,0],X[:,1], c=y, cmap='coolwarm')
```
