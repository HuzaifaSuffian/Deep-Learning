```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
x1_input = [1,0,1,0]
x2_input = [1,0,0,1]
output1 = [1,0,0,0]
```

```python
x3_input = [1,0,1,0]
x4_input = [1,0,0,1]
output2 = [1,0,1,1]
```

```python
x5_input = [1,0,1,0]
x6_input = [1,0,0,1]
output3 = [0,0,1,1]
```

```python
and_df = pd.DataFrame({'x1':x1_input,'x2':x2_input,'output':output1})
or_df = pd.DataFrame({'x3':x3_input,'x4':x4_input,'output':output2})
xor_df = pd.DataFrame({'x5':x5_input,'x6':x6_input,'output':output3})
```

```python
plt.scatter(and_df.iloc[:,0],and_df.iloc[:,1],c=and_df.iloc[:,-1])
plt.title('And DataFrame')
```

```python
plt.scatter(or_df.iloc[:,0],or_df.iloc[:,1],c=or_df.iloc[:,-1])
plt.title('Or DataFrame')
```

```python
plt.scatter(xor_df.iloc[:,0],xor_df.iloc[:,1],c=xor_df.iloc[:,-1])
plt.title('XOR DataFrame')
```

```python
from sklearn.linear_model import Perceptron
perc1 = Perceptron()
perc2 = Perceptron()
perc3 = Perceptron()
```

```python
clf1 = perc1.fit(and_df.iloc[:,:2],and_df.iloc[:,-1])
```

```python
clf2 = perc2.fit(or_df.iloc[:,:2],or_df.iloc[:,-1])
```

```python
clf3 = perc3.fit(xor_df.iloc[:,:2],xor_df.iloc[:,-1])
```

```python
m1 = -(clf1.coef_[0][0]/clf1.coef_[0][1])
b1 = -(clf1.intercept_/clf1.coef_[0][1])
```

```python
m2 = -(clf2.coef_[0][0]/clf2.coef_[0][1])
b2 = -(clf2.intercept_/clf2.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
plt.plot(x_input, y_input)
plt.scatter(and_df.iloc[:,0],and_df.iloc[:,1],c=and_df.iloc[:,-1])
plt.title('And DataFrame')
```

```python
x1_input = np.linspace(-3,3,100)
y1_input = m2*x1_input + b2
plt.plot(x1_input, y1_input)
plt.scatter(or_df.iloc[:,0],or_df.iloc[:,1],c=or_df.iloc[:,-1])
plt.title('Or DataFrame')
```

```python
plot_decision_regions(X=xor_df.iloc[:,:2].values, y=xor_df.iloc[:,-1].values, clf=clf3)
```
