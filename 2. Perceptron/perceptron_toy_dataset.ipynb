```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples = 200, n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,class_sep=5,random_state=45)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.linear_model import Perceptron
perc = Perceptron()
perc.fit(X,y)
print(perc.coef_)
print(perc.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y,clf=perc,legend=2)
```
